## 1. Retrieve Data from MongoDB
Retreive all documents from MongoDB and convrt it to a `pandas.DataFrame`

```python
import pymongo

client = pymongo.MongoClient()
```

```python
import pandas as pd
client = pymongo.MongoClient()
db = client.get_database("states")
collection = db.get_collection("states")
data = list(collection.find())
df_s = pd.DataFrame.from_records(data)    
df_s.drop('_id', axis=1, inplace=True)
df_s.head()
```

## 2 Correlation: cases and deaths

Let's check the correlation between cases and deaths in each state over time


```python
import plotly.express as px
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
fig = px.scatter_matrix(df_s,
    dimensions=["deaths", "cases"],
    color="state", symbol='state')

fig.update_layout(
    title='Scatter matrix of covid cases and deaths
    dragmode='select',
    width=720,
    height=720,
    hovermode='closest',
)

fig.show()
```
